```python
from bs4 import BeautifulSoup
import pandas as pd
import requests

# List of URLs to scrape
url_list = [
     # islam:
    'http://www.ufcstats.com/fighter-details/275aca31f61ba28c',
    # charles:
    'http://www.ufcstats.com/fighter-details/07225ba28ae309b6',
    # gaethje:
    'http://www.ufcstats.com/fighter-details/9e8f6c728eb01124',
    # Dustin:
    'http://www.ufcstats.com/fighter-details/029eaff01e6bb8f0',
    # Beneil:
    'http://www.ufcstats.com/fighter-details/08af939f41b5a57b',
    # Chandler:
    'http://www.ufcstats.com/fighter-details/4b93a88f3b1de35b',
    # Rafael:
    'http://www.ufcstats.com/fighter-details/c814b4c899793af6',
    # Gamrot:
    'http://www.ufcstats.com/fighter-details/72db2a14ffa73ece',
    #Tsarukyan:
    'http://www.ufcstats.com/fighter-details/eae48ff31db420c2',
    # Hooker:
    'http://www.ufcstats.com/fighter-details/193b9d1858bc4df3',
    # Grant:
    'http://www.ufcstats.com/fighter-details/99bd51917728c25d',
    # Dos Anjos:
    'http://www.ufcstats.com/fighter-details/6a2f7c151031653d',
    # Jalin :
    'http://www.ufcstats.com/fighter-details/4d8fa64202cee4c1',
    # Moicano:
    'http://www.ufcstats.com/fighter-details/b6452706b373eea1',
    # frevola:
    'http://www.ufcstats.com/fighter-details/dccb63727f2f5f74',
    # Ferreria:
    'http://www.ufcstats.com/fighter-details/f53d4f21d1b5f2dc'
]

# Create an empty list to store DataFrames for each URL
dfs = []

for url in url_list:
    # Make an HTTP request to fetch the HTML content
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        html_content = response.text

        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')

        # Extract fighter's name
        fighter_name_elem = soup.find("span", class_="b-content__title-highlight")
        if fighter_name_elem:
            fighter_name = fighter_name_elem.get_text(strip=True)

        # Extract career statistics
        career_stats = soup.find("div", class_="b-list__info-box-left")
        if career_stats:
            stats_items = career_stats.find_all("li", class_="b-list__box-list-item")
            data = {'Fighter Name': fighter_name}  # Include fighter's name in the data dictionary
            for item in stats_items:
                title = item.find("i", class_="b-list__box-item-title").get_text(strip=True)
                value = item.get_text(strip=True).replace(title, '', 1).strip()
                data[title] = value

        # Extract additional statistics
        additional_stats = soup.find("div", class_="b-list__info-box-right")
        if additional_stats:
            stats_items = additional_stats.find_all("li", class_="b-list__box-list-item")
            for item in stats_items:
                title = item.find("i", class_="b-list__box-item-title").get_text(strip=True)
                value = item.get_text(strip=True).replace(title, '', 1).strip()
                data[title] = value

        # Create a DataFrame from the extracted data
        df = pd.DataFrame([data])

        # Rename columns for readability
        df = df.rename(columns={
            'SLpM:': 'SLpM',
            'Str. Acc.:': 'Str_Acc',
            'SApM:': 'SApM',
            'Str. Def:': 'Str_Def',
            'TD Avg.:': 'TD_Avg',
            'TD Acc.:': 'TD_Acc',
            'TD Def.:': 'TD_Def',
            'Sub. Avg.:': 'Sub_Avg'
        })

        # Append the DataFrame to the list of DataFrames
        dfs.append(df)

    else:
        print(f"Failed to fetch data from URL: {url}")

# Concatenate all DataFrames in the list
final_df = pd.concat(dfs, ignore_index=True)

# Print the final DataFrame
print(final_df)

# Assuming df is the DataFrame you want to save
data_to_save = final_df

# Specify the file path where you want to save the CSV
csv_file_path = "ufc_lightweight_data_strike_TD_data.csv"  # Use an appropriate path

# Save the DataFrame to a CSV file
data_to_save.to_csv(csv_file_path, index=False)
```
